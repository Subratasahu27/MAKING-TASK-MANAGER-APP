```python
# Define a Task class to store task information
class Task:
  def __init__(self, description):
    self.description = description
    self.completed = False  # Optional: Track task completion status

# Create an empty list to store tasks
tasks = []

def add_task():
  """Prompts user for task description and adds it to the list"""
  description = input("Enter task description: ")
  new_task = Task(description)
  tasks.append(new_task)
  print("Task added successfully!")

def list_tasks():
  """Displays all tasks in the list"""
  if not tasks:
    print("There are no tasks to display.")
    return
  for i, task in enumerate(tasks):
    print(f"{i+1}. {task.description}")

def remove_task():
  """Displays tasks and allows user to remove one"""
  if not tasks:
    print("There are no tasks to remove.")
    return
  list_tasks()
  index = int(input("Enter the number of the task to remove: ")) - 1
  if index < 0 or index >= len(tasks):
    print("Invalid task number.")
    return
  del tasks[index]
  print("Task removed successfully!")

# Main program loop
while True:
  print("\nTask Management App")
  print("1. Add Task")
  print("2. List Tasks")
  print("3. Remove Task")
  print("4. Exit")
  choice = input("Enter your choice (1-4): ")

  if choice == '1':
    add_task()
  elif choice == '2':
    list_tasks()
  elif choice == '3':
    remove_task()
  elif choice == '4':
    print("Exiting...")
    break
  else:
    print("Invalid choice. Please try again.")
```

```python
# Define a Task class
class Task:
  def __init__(self, description, priority):
    self.description = description
    self.priority = priority  # Can be "high", "medium", or "low"
    self.completed = False  # Optional: Track completion status

# Function to display all tasks
def list_tasks(tasks):
  if not tasks:
    print("No tasks added yet!")
    return
  print("Tasks:")
  for i, task in enumerate(tasks):
    print(f"{i+1}. {task.description} ({task.priority})")

# Function to add a new task
def add_task(tasks):
  description = input("Enter task description: ")
  priority = input("Enter priority (high, medium, low): ").lower()
  if priority not in ("high", "medium", "low"):
    print("Invalid priority. Please choose high, medium, or low.")
    return
  tasks.append(Task(description, priority))
  print("Task added successfully!")

# Function to prioritize a task
def prioritize_task(tasks):
  if not tasks:
    print("No tasks to prioritize!")
    return
  list_tasks(tasks)
  index = int(input("Enter the number of the task to prioritize: ")) - 1
  if index < 0 or index >= len(tasks):
    print("Invalid task index.")
    return
  new_priority = input("Enter new priority (high, medium, low): ").lower()
  if new_priority not in ("high", "medium", "low"):
    print("Invalid priority. Please choose high, medium, or low.")
    return
  tasks[index].priority = new_priority
  print("Task priority updated!")

# Main program loop
tasks = []
while True:
  print("\nTask Manager")
  print("1. List Tasks")
  print("2. Add Task")
  print("3. Prioritize Task")
  print("4. Exit")
  choice = input("Enter your choice: ")

  if choice == '1':
    list_tasks(tasks)
  elif choice == '2':
    add_task(tasks)
  elif choice == '3':
    prioritize_task(tasks)
  elif choice == '4':
    break
  else:
    print("Invalid choice. Please try again.")

print("Exiting Task Manager...")
```

```python
# Define a task structure
class Task:
  def __init__(self, description, priority="medium"):
    self.description = description
    self.priority = priority

# Define a list to store tasks
tasks = []

def add_task():
  description = input("Enter task description: ")
  priority = input("Enter priority (high, medium, low): ").lower()
  new_task = Task(description, priority)
  tasks.append(new_task)
  print("Task added successfully!")

def remove_task():
  if not tasks:
    print("No tasks to remove!")
    return
  
  for i, task in enumerate(tasks):
    print(f"{i+1}. {task.description} ({task.priority})")
  
  choice = int(input("Enter task number to remove (or 0 to cancel): ")) - 1
  if 0 <= choice < len(tasks):
    del tasks[choice]
    print("Task removed successfully!")
  else:
    print("Invalid task number!")

def list_tasks():
  if not tasks:
    print("No tasks to list!")
    return
  
  for i, task in enumerate(tasks):
    print(f"{i+1}. {task.description} ({task.priority})")

def prioritize_task():
  if not tasks:
    print("No tasks to prioritize!")
    return
  
  for i, task in enumerate(tasks):
    print(f"{i+1}. {task.description} ({task.priority})")
  
  choice = int(input("Enter task number to prioritize (or 0 to cancel): ")) - 1
  if 0 <= choice < len(tasks):
    new_priority = input("Enter new priority (high, medium, low): ").lower()
    tasks[choice].priority = new_priority
    print("Task priority updated successfully!")
  else:
    print("Invalid task number!")

def main_menu():
  print("\nTask Management App")
  print("1. Add Task")
  print("2. Remove Task")
  print("3. List Tasks")
  print("4. Prioritize Task")
  print("5. Exit")

  choice = input("Enter your choice (1-5): ")
  
  if choice == '1':
    add_task()
  elif choice == '2':
    remove_task()
  elif choice == '3':
    list_tasks()
  elif choice == '4':
    prioritize_task()
  elif choice == '5':
    exit()
  else:
    print("Invalid choice!")

# Main program loop
while True:
  main_menu()
```
